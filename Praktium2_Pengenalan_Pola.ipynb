##Inisialisasi Library

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

##Pembangkitan data untuk setiap kelas

```python
# Dua Kelas (Asumsi menggunakan Gaussian)
np.random.seed(42)

# Membangkitkan Data kelas 1
mu1, sigma1 = 2, 0.8
x1 = np.random.normal(mu1, sigma1, 100)

# Membangkitkan Data kelas 2
mu2, sigma2 = 5, 1.0
x2 = np.random.normal(mu1, sigma1, 100)
```

##Menentukan Posterior Probability

```python
# Posterior Probability
P_w1 = 0.4
P_w2 = 0.6
```

##Membuat fungsi likelihood

```python
def likelihood(x, mu, sigma):
  return norm.pdf(x, mu, sigma)
```

##Membuat fungsi posterior probabability

```python
# Fungsi posterior probability dengan pendekatan konsep bayes
# Mencari peluang baer menggunakan fungsi likelihood dengan fungsi bayes

def posterior(x):
  p1 = likelihood(x, mu1, sigma1) * P_w1
  p2 = likelihood(x, mu2, sigma2) * P_w2
  return p1, p2
```

##Membuat fungsi klasifikasi berdasarkan nilai posterior probability

```python
# Minimum Error Rule
# Fungsi untuk klasifikasi dengan memanggil fungsi posterior probability

def classify(x):
  p1, p2 = posterior(x)
  return np.where(p1 >p2, 1, 2)
```

##Pembangkitan visualisasi data

```python
x_range = np.linspace(-1, 9, 100)

p1 = likelihood(x_range, mu1, sigma1) * P_w1
p2 = likelihood(x_range, mu2, sigma2) * P_w2

print(p1)
print("\n")
print(p2)

plt.figure()
plt.plot(x_range, p1, label='Kelas 1')
plt.plot(x_range, p2, label='Kelas 2')
plt.xlabel('x')
plt.ylabel('Posterior (unnormalized)')
plt.legend()
plt.title('Bayes Minimum Error Classification')
plt.show()
```

##Klasifikasi dan Menghitung Nilai Error

```python
# Gabungkan data
data = np.concatenate([x1, x2])
true_labels = np.array([1] * len(x1) + [2] * len(x2))

# Klasifikasi data
pred_label = classify(data)
print(pred_label)

# Menghitung error
error_rate = np.mean(pred_label != true_labels)
print("Error Rate:", error_rate)
```
